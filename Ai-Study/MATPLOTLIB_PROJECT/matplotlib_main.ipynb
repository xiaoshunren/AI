```python
import matplotlib.pyplot as mp # 数据可视化
# import pandas as pd # 序列高级函数
# import scipy as sc # 科学计算
import numpy as np # 基础值算法

# 进行线性拆分1000个点

x = np.linspace(-np.pi,np.pi,1000)
print(x,x.shape)



sinx = np.sin(x)
cosx = np.cos(x)
# 绘制余弦曲线 cos(x)/2
cosx = np.cos(x) / 2

# 设置坐标轴的范围
# mp.xlim(0,np.pi)
# mp.ylim(0,1)

# x/y轴 值列表 文本列表
# -π,-π/2,0,π/2,π
# frac{}{} 分数 {分子}{分母}
vals=[-np.pi,-np.pi/2,0, np.pi/2,np.pi]
 #1 texts=['-π','-π/2','0','π/2','π']
texts=[r'$-\pi$',r'$-\frac{\pi}{2}$','0',r'$\frac{\pi}{2}$',r'$-\pi$']
mp.xticks(vals,texts)

# 修改坐标轴
ax = mp.gca()
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))

mp.yticks([-1.0,-0.5,0.5,1.0])

# 线的样式，粗细，颜色,标签
mp.plot(x,sinx, linestyle = '--', linewidth=2, color='orangered', alpha=0.8,label=r'$y=sin(x)$')
mp.plot(x,cosx, linestyle ='-.', linewidth=2, color='dodgerblue', alpha = 0.9,label=r'$y=\frac{1}{2}cos(x)$')

# 绘制特殊点
pointx = [np.pi/2,np.pi/2]
pointy =[1,0]
mp.scatter(pointx,pointy,
           marker='o',s=70,color='red',
           label = 'sample points',zorder=3) # 图层优先度zorder

# 图例
mp.legend()
mp.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 分簇散点图
# 分簇散点图Swarm Plot 通常在 x 轴上显示分类变量，在y轴上显示数值变量，
# 并沿着每个类别显示单个数据点。可以直观地看到类别内数据点的分布和密度。
tips = sns.load_dataset("tips")
tips.head()
plt.figure(figsize=(8,5))
sns.swarmplot(x="day",y="total_bill",data=tips,palette="viridis")
plt.title("Total Bill Distribution by Day of the Week") # 按星期分布
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill($)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 按周来统计小费数量
tips = sns.load_dataset("tips")
tips.head()
plt.figure(figsize=(8,5))
sns.countplot(x='day',data=tips,palette="Set3")
plt.title("Count of Tips by Day of the week" )
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 点阵图
# 点阵图Point Plot通常在 x 轴上显示分类变量，在 y 轴上显示数值变量，
# 可以深入了解每类数据的分布和中心点倾向。
tips = sns.load_dataset("tips")
plt.figure(figsize=(8,5))
sns.pointplot(x="day",y="total_bill",data=tips,ci="sd",palette="pastel")
plt.title("Average Total Bill by Day of The week") # 按星期计算的平均小费
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill($)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load the "tips" dataset
tips = sns.load_dataset("tips")
# sns.violinplot 绘制小提琴图。
# 分类变量通常控制组或类别，而数值变量则用于在每个类别中创建小提琴图。
# 小提琴图中某一点的宽度表示该值所在数据点的密度。
plt.figure(figsize=(8, 5))
sns.violinplot(x="day", y="total_bill",data=tips,palette="Set2")
plt.title("Total Bill Distribution by Day of the Week") # 按星期分布的总账单
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load the "tips" dataset
tips = sns.load_dataset("tips")
# 二维变量回归图
# 回归图(Regression Plot)的重点是突出两个数字变量之间的关系
# 自变量（通常在 x 轴上）和因变量（在 y 轴上）。单个数据点以点的形式显示，回归图的中心元素是回归线或曲线，它代表了描述变量之间关系的最佳拟合数学模型。
plt.figure(figsize=(8, 5))
sns.regplot(x="total_bill", y="tip", data=tips, scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.title("Regression Plot of Total Bill vs. Tip") # 总账单与小费的回归图
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load the "tips" dataset
tips = sns.load_dataset("tips")
# 联合分布图(Joint Plot)
# 结合了散点图、直方图和密度图来直观显示两个数值变量之间的关系。
# 散点图是联合图的核心元素，它将两个变量的数据点相对应地显示出来，
# 沿着散点图的 x 轴和 y 轴，每个变量都有直方图或核密度估计图。这些边际图分别显示每个变量的分布情况。
sns.jointplot(x="total_bill", y="tip", data=tips, kind="scatter")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load the "tips" dataset
tips = sns.load_dataset("tips")
# 六边形图(Hexbin plot)
# 它将数据点分组到六边形箱中，能够更有效地直观显示数据密度和模式。
# 当单个点的散点图变得过于拥挤且难以解读时，六边形分隔图对于这类大型数据集尤为重要
plt.figure(figsize=(8, 5))
sns.jointplot(x="total_bill", y="tip",kind='hex', data=tips, gridsize=15, cmap="Blues")
plt.title("Hexbin Plot of Total Bill vs. Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load the "tips" dataset
tips = sns.load_dataset("tips")
# 通过关系图，大家可以直观地看到两个数值变量之间的关系，以及其他分类或数值维度。
# 一般使用函数sns.relplot()进行绘制，可以使用色调参数根据分类变量为数据点着色，
# 使用大小参数根据数值变量改变标记大小，使用样式参数根据分类变量区分标记或线条。
sns.relplot(x="total_bill", y="tip", data=tips, hue="time", style="sex", size="size", palette="Set2", height=6)
plt.title("Relational Scatter Plot for Tips Dataset") # 小费数据集的关系散点图
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.savefig('relplot_with_mv.png')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load the "tips" dataset
tips = sns.load_dataset("tips")
# 为不同的天创建直方图
# 面网格图是Seaborn的一项功能，允许创建一个子网格，每个子网格代表数据集中的一个不同子集。
# 通过这种方式，面网格图可用于比较不同类别中多个变量之间的关系。
g = sns.FacetGrid(tips, col="day", height=4, aspect=1.2)
g.map(sns.histplot, "total_bill", kde=True)
g.set_axis_labels("Total Bill ($)", "Frequency") # 频率
g.set_titles(col_template="{col_name} Day")
plt.savefig('facet_grid_hist_plot.png')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Load the "iris" dataset
iris = sns.load_dataset("iris")
# 配对图(Pair Plot)提供了一个由散点图和直方图组成的网格，
# 其中每个图都显示了两个变量之间的关系，因此也被称为配对图或散点图矩阵
sns.set(style="ticks") #设置Seaborn的绘图样式为“ticks”，这会改变图形的整体外观，使其更美观
sns.pairplot(iris, hue="species", markers=["o", "s", "D"])
plt.show()
```
